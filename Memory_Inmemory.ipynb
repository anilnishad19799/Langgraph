```python
from langchain.chat_models import init_chat_model
from langgraph.graph import StateGraph, MessagesState, START
from langgraph.checkpoint.memory import InMemorySaver

# Initialize the chat model
model = init_chat_model(model="openai:gpt-4o")

# Use in-memory checkpointer
checkpointer = InMemorySaver()

# Define the model invocation node
def call_model(state: MessagesState):
    response = model.invoke(state["messages"])
    return {"messages": response}

# Build the state graph
builder = StateGraph(MessagesState)
builder.add_node("call_model", call_model)
builder.add_edge(START, "call_model")

# Compile the graph with in-memory checkpointing
graph = builder.compile(checkpointer=checkpointer)

# Configuration for thread ID
config = {
    "configurable": {
        "thread_id": "1"
    }
}


system_prompt="""

  You are a female assistant named Naayra at a Ecommerce shop named Indiamart.
  For any queries asked respond in a professional and polite tone

"""


# First interaction
initial_state = {
    "messages": [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": "hi! I'm Ariki nagasawa"}
    ]
}



for chunk in graph.stream(
    # {"messages": [{"role": "user", "content": "hi! I'm bob"}]},
    initial_state,
    config,
    stream_mode="values"
):
    chunk["messages"][-1].pretty_print()

```
